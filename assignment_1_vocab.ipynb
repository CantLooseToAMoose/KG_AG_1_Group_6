As we can see here each nutrition has a recipe but not each recipe has a nutrition so we can merge nutrition on recipe we will be left with NaNs for the recipes without nutrition but that is completely fine knowledge graphs can handle this. 
Now for the construction of the KG we need some way to link recipes to their nutrition however nutrition can have more than one entry for a dish, but these are not uniquely traceable back to their original recipe thus I cant identify which nutrition of a dish belongs to a specific recipe, therefore I assume each dish has the same nutrition but can have several recipes otherwise it would be impossible or too difficult to identify wich belong to which. Therefore I will also drop duplicate entries in the nutrition table to be inline with this assumption of only having one nutrition for a dish. After dropping I will perform a left outerjoin on recipes with nutrition to make one table for ease of knowledge graph construction.

Now one more standardization is I will put the ingredients and quatities together, one downside is we are not given units for the amount of an ingredient so this will be missing in the data but one can most likely infer from the ingredient what unit it should be if there is a NaN value it will simple be taken over so we might have just the ingredient or only the amount.

```python
import re


def sanitize_name(name):
    # Lowercase and strip whitespace
    name = name.strip().lower()
    # Replace spaces with underscores
    name = re.sub(r'\s+', '_', name)
    # Optionally remove characters that aren't alphanumeric or underscores/hyphens
    name = re.sub(r'[^\w-]', '', name)
    return name


recipes_with_nutrition['Name'] = recipes_with_nutrition['Name'].apply(sanitize_name)
```

```python
def parse_r_vector(r_vector_str):
    """Removes the leading 'c(' and trailing ')' and splits the string into a list."""
    # If the input is not a string, return an empty list
    if not isinstance(r_vector_str, str):
        return []
    cleaned = re.sub(r"^c\(|\)$", "", r_vector_str)
    # Split on comma+space and remove any remaining quotes and whitespace
    items = [item.strip().strip('"').strip("'") for item in cleaned.split(", ")]
    return items


def combine_ingredients(row):
    """Combine quantities and ingredients for a given row."""
    # Since the columns have already been parsed, they are now lists
    quantities = row['RecipeIngredientQuantities']
    ingredients = row['RecipeIngredientParts']
    return [f"{qty} {ing}" for qty, ing in zip(quantities, ingredients)]


# Edit: Johannes I will create a new Dataframe here from the recipe_with_nutritions thing

# convert into correct format from r vector to list
# recipes_with_nutrition_copy = recipes_with_nutrition.copy()
# recipes_with_nutrition_copy['RecipeIngredientQuantities'] = recipes_with_nutrition_copy[
#     'RecipeIngredientQuantities'].apply(
#     parse_r_vector)
# recipes_with_nutrition_copy['RecipeIngredientParts'] = recipes_with_nutrition_copy['RecipeIngredientParts'].apply(
#     parse_r_vector)

# filter out rows where quantity and ingredients do not match
# recipes_with_nutrition_copy = recipes_with_nutrition_copy[
#     recipes_with_nutrition_copy['RecipeIngredientQuantities'].apply(len) == recipes_with_nutrition_copy[
#         'RecipeIngredientParts'].apply(len)]

# explode the list to get unique rows
# recipes_with_nutrition_copy = recipes_with_nutrition_copy.explode(
# ['RecipeIngredientQuantities', 'RecipeIngredientParts'])

# Rename columns
# recipes_with_nutrition_copy = recipes_with_nutrition_copy.rename(
# columns={'RecipeIngredientQuantities': 'Quantity', 'RecipeIngredientParts': 'Ingredient'})

# recipes_ingredients_dataframe = recipes_with_nutrition_copy[["RecipeId", 'Quantity', 'Ingredient']]

# Actually make sure that formatting is consisent by stripping leading and trailing whitespaces and lower casing

# recipes_ingredients_dataframe["Ingredient"] = recipes_ingredients_dataframe["Ingredient"].str.lower().str.strip()

# Apply the function to each row to create a new column "Ingredients"
recipes_with_nutrition['RecipeIngredientQuantities'] = recipes_with_nutrition['RecipeIngredientQuantities'].apply(
    parse_r_vector)
recipes_with_nutrition['RecipeIngredientParts'] = recipes_with_nutrition['RecipeIngredientParts'].apply(parse_r_vector)
recipes_with_nutrition['Keywords'] = recipes_with_nutrition['Keywords'].apply(parse_r_vector)
recipes_with_nutrition['Ingredients'] = recipes_with_nutrition.apply(combine_ingredients, axis=1)

# Drop the now unnecessary columns
# recipes_with_nutrition.drop(columns=['RecipeIngredientQuantities', 'RecipeIngredientParts'], inplace=True)

# Save into .csv
# recipes_ingredients_dataframe.to_csv("Ingredients.csv", index=False)
# Clear from Cache
recipes_with_nutrition_copy = None
```

```python

def parse_cuisines(cuisine_str):
    """
    Converts a string of cuisines into a list of cuisines.
    
    Parameters:
        cuisine_str (str): A string containing one or more cuisines,
                           separated by commas if multiple.
    
    Returns:
        list: A list of cuisines (each cuisine is stripped of extra whitespace).
    """
    # Check for missing or non-string values
    if not isinstance(cuisine_str, str) or not cuisine_str.strip():
        return []

    # Split on comma and strip whitespace from each cuisine
    return [cuisine.strip() for cuisine in cuisine_str.split(',')]


restaurants['Cuisine_List'] = restaurants['Cuisines'].apply(parse_cuisines)
```

```python
import rdflib
from rdflib import Graph, Namespace, URIRef, Literal, RDF, RDFS, OWL
from rdflib.namespace import DCTERMS, XSD, RDF, SDO, RDFS, SKOS
```

```python
ex = Namespace("http://kg-course/food-nutrition/")

# Create Graph
gr = rdflib.Namespace("http://purl.org/goodrelations/v1#")
dbo = rdflib.Namespace("http://dbpedia.org/ontology/")
dbr = rdflib.Namespace("http://dbpedia.org/resource/")

g = Graph()
g.bind("dbo", dbo)
g.bind("dbr", dbr)
g.bind("rdf", RDF)
g.bind("rdfs", RDFS)
g.bind("gr", gr)
g.bind("owl", OWL)
g.bind("xsd", XSD)
g.bind("schema", SDO)
g.bind("skos", SKOS)
g.bind("ex", ex)

#The following classes and properties already exist in schema.org ontologies thus I wont be redifing them here however I will list them as they will be used
'''
classes
schema:Recipe
schema:Restaurant
OWL:Thing
schema:Review
schema:rating
schema:Person



properties
#for recipes
we drop recipeServings as its essentially equivalent to recipeYield.
name
recipeInstructions
cookTime
prepTime
image
datePublished
recipeYield
recipeCategory
recipeIngredient

#for nutrition
calories	
fatContent	
saturatedFatContent
cholesterolContent	
sodiumContent	
carbohydrateContent	
fiberContent	
sugarContent	
proteinContent

#for restaurant 
servesCuisine
location
address
legalName
lonngitude
latitude
currenciesAccepted
priceRange
we choose to combine avergacostof2 and currency for priceRange as text.

#for reviews
review
itemReviewed
author
ratingValue
datePublished
dateSubmitted
reviewBody
alternateName

'''

#Defining all the classes
#dish
g.add((ex.Dish, RDF.type, RDFS.Class))
g.add((ex.Dish, RDFS.label, Literal('Dish')))
g.add((ex.Dish, RDFS.comment, Literal('Something consumed when famished.')))
g.add((ex.Dish, RDFS.subClassOf, OWL.Thing))

#cuisine
g.add((ex.Cuisine, RDF.type, RDFS.Class))
g.add((ex.Cuisine, RDFS.label, Literal('Cuisine')))
g.add((ex.Cuisine, RDFS.comment, Literal('A type of cuisine typically belonging to a restaurant or recipe')))
g.add((ex.Cuisine, RDFS.subClassOf, OWL.Thing))

#NutritionInformation
g.add((ex.NutritionInfo, RDF.type, RDFS.Class))
g.add((ex.NutritionInfo, RDFS.label, Literal('NutritionInfo')))
g.add((ex.NutritionInfo, RDFS.comment, Literal("Nutritional information for a dish.")))
g.add((ex.NutritionInfo, OWL.equivalentClass, SDO.NutritionInformation))

#Ingredient
g.add((ex.Ingredient, RDF.type, RDFS.Class))
g.add((ex.Ingredient, RDFS.label, Literal('Ingredient')))
g.add((ex.Ingredient, RDFS.comment, Literal('Ingredient used in a recipe')))
g.add((ex.Ingredient, RDFS.subClassOf, OWL.Thing))

#Defining all the properties
#nutrition
g.add((ex.nutrition, RDF.type, RDF.Property))
g.add((ex.nutrition, RDFS.domain, ex.Dish))
g.add((ex.nutrition, RDFS.range, ex.NutritionInfo))
g.add((ex.nutrition, RDFS.label, Literal('nutrition')))
g.add((ex.nutrition, RDFS.comment, Literal('Links a dish to its nutrition info.')))
g.add((ex.nutrition, SKOS.closeMatch, SDO.nutrition))

#hasRecipe
g.add((ex.hasRecipe, RDF.type, RDF.Property))
g.add((ex.hasRecipe, RDFS.domain, ex.Dish))
g.add((ex.hasRecipe, RDFS.range, SDO.Recipe))
g.add((ex.hasRecipe, RDFS.label, Literal('hasRecipe')))
g.add((ex.hasRecipe, RDFS.comment, Literal('Links a dish to its recipe, a dish may have several recipes.')))

#inCOuntry
g.add((ex.inCountry, RDF.type, RDF.Property))
g.add((ex.inCountry, RDFS.domain, SDO.Restaurant))
g.add((ex.inCountry, RDFS.range, XSD.string))
g.add((ex.inCountry, RDFS.label, Literal('InCountry')))
g.add((ex.inCountry, RDFS.comment, Literal('Links a restaurant to its country of origin.')))
g.add((ex.inCountry, SKOS.closeMatch, SDO.addressCountry))

#hasTableBooking
#hasDelivery
g.add((ex.hasTableBooking, RDF.type, RDF.Property))
g.add((ex.hasTableBooking, RDFS.domain, SDO.Restaurant))
g.add((ex.hasTableBooking, RDFS.range, XSD.boolean))
g.add((ex.hasTableBooking, RDFS.label, Literal('hasTableBooking')))
g.add((ex.hasTableBooking, RDFS.comment, Literal('Links a restaurant to its ability to be able to book a table.')))

#hasDelivery
g.add((ex.hasDelivery, RDF.type, RDF.Property))
g.add((ex.hasDelivery, RDFS.domain, SDO.Restaurant))
g.add((ex.hasDelivery, RDFS.range, XSD.boolean))
g.add((ex.hasDelivery, RDFS.label, Literal('hasDelivery')))
g.add((ex.hasDelivery, RDFS.comment,
       Literal('Links a restaurant to its ability to deliver food to your house via online order.')))

#isDeliveringNow
g.add((ex.isDeliveringNow, RDF.type, RDF.Property))
g.add((ex.isDeliveringNow, RDFS.domain, SDO.Restaurant))
g.add((ex.isDeliveringNow, RDFS.range, XSD.boolean))
g.add((ex.isDeliveringNow, RDFS.label, Literal('isDeliveringNow')))
g.add((ex.isDeliveringNow, RDFS.comment, Literal('Links a restaurant to the fact if its delivering now.')))

#switchToOrderMenu
g.add((ex.switchToOrderMenu, RDF.type, RDF.Property))
g.add((ex.switchToOrderMenu, RDFS.domain, SDO.Restaurant))
g.add((ex.switchToOrderMenu, RDFS.range, XSD.boolean))
g.add((ex.switchToOrderMenu, RDFS.label, Literal('switchToOrderMenu')))
g.add(
    (ex.switchToOrderMenu, RDFS.comment, Literal('The availibility to have the ability to switch to the order menu.')))

#avgRating
g.add((ex.avgRating, RDF.type, RDF.Property))
g.add((ex.avgRating, RDFS.domain, SDO.Restaurant))
g.add((ex.avgRating, RDFS.range, XSD.double))
g.add((ex.avgRating, RDFS.label, Literal('avgRating')))
g.add((ex.avgRating, RDFS.comment, Literal('The average rating given to a restaurant by customers.')))
g.add((ex.avgRating, SKOS.closeMatch, SDO.ratingValue))

#votes
g.add((ex.votes, RDF.type, RDF.Property))
g.add((ex.votes, RDFS.domain, SDO.Restaurant))
g.add((ex.votes, RDFS.range, XSD.integer))
g.add((ex.votes, RDFS.label, Literal('votes')))
g.add((ex.votes, RDFS.comment, Literal('Amount of reviews received')))
g.add((ex.votes, SKOS.closeMatch, SDO.reviewCount))

#numberOfDishes
g.add((ex.numberOfDishes, RDF.type, RDF.Property))
g.add((ex.numberOfDishes, RDFS.domain, SDO.Restaurant))
g.add((ex.numberOfDishes, RDFS.range, XSD.integer))
g.add((ex.numberOfDishes, RDFS.label, Literal('numberOfDishes')))
g.add((ex.numberOfDishes, RDFS.comment, Literal('Amount of unque dishes served by a restaurant.')))

#hasCuisine
g.add((ex.hasCuisine, RDF.type, RDF.Property))
g.add((ex.hasCuisine, RDFS.domain, SDO.Restaurant))
g.add((ex.hasCuisine, RDFS.range, ex.Cuisine))
g.add((ex.hasCuisine, RDFS.label, Literal('hasCuisine')))
g.add((ex.hasCuisine, RDFS.comment, Literal('Links a restaurant a cuisine it servers.')))
g.add((ex.hasCuisine, SKOS.closeMatch, SDO.servesCuisine))

#containsIngredient
g.add((ex.containsIngredient, RDF.type, RDF.Property))
g.add((ex.containsIngredient, RDFS.domain, SDO.Recipe))
g.add((ex.containsIngredient, RDFS.range, ex.Ingredient))
g.add((ex.containsIngredient, RDFS.label, Literal('containsIngredient')))
g.add((ex.containsIngredient, RDFS.comment, Literal('Links an ingredient to its recipe.')))
g.add((ex.containsIngredient, SKOS.closeMatch, SDO.recipeIngredient))


```

```python
def safe_uri_local(s: str) -> str:
    """
    Convert 'milk & cookies' â†’ 'milk_and_cookies'
    """
    return re.sub(r'[^a-zA-Z0-9_]+', '_', s.strip()).lower()


def add_triples(g=None, recipes_with_nutrition=None, reviews=None, restaurants=None, amount=10000):
    # lets start adding everything to the graph please keep my comments
    dish_instances = {}  # keep track of dishes I've seen
    ingredient_instances = {}

    if recipes_with_nutrition is not None:
        for index, row in recipes_with_nutrition.iloc[:amount].iterrows():
            # create dishes
            if not pd.isna(row['Name']):
                dish = row['Name']
                if dish not in dish_instances:
                    dish_uri = ex[dish]
                    dish_instances[dish] = dish_uri

                    # create dish
                    g.add((dish_uri, RDF.type, ex.Dish))
                    # add nutritionInformation
                    nutrition_info_uri = ex[f'{dish}_nutritionInfo']
                    g.add((nutrition_info_uri, RDF.type, ex.NutritionInfo))
                    g.add((dish_uri, ex.nutrition, nutrition_info_uri))
                    # add all information to nutritionInformation
                    if not pd.isna(row['Calories']):
                        g.add((nutrition_info_uri, SDO.calories, Literal(row['Calories'], datatype=XSD.float)))

                    if not pd.isna(row['FatContent']):
                        g.add((nutrition_info_uri, SDO.fatContent, Literal(row['FatContent'], datatype=XSD.float)))

                    if not pd.isna(row['SaturatedFatContent']):
                        g.add((nutrition_info_uri, SDO.saturatedFatContent,
                               Literal(row['SaturatedFatContent'], datatype=XSD.float)))

                    if not pd.isna(row['CholesterolContent']):
                        g.add((nutrition_info_uri, SDO.cholesterolContent,
                               Literal(row['CholesterolContent'], datatype=XSD.float)))

                    if not pd.isna(row['SodiumContent']):
                        g.add(
                            (nutrition_info_uri, SDO.sodiumContent, Literal(row['SodiumContent'], datatype=XSD.float)))

                    if not pd.isna(row['CarbohydrateContent']):
                        g.add((nutrition_info_uri, SDO.carbohydrateContent,
                               Literal(row['CarbohydrateContent'], datatype=XSD.float)))

                    if not pd.isna(row['FiberContent']):
                        g.add((nutrition_info_uri, SDO.fiberContent, Literal(row['FiberContent'], datatype=XSD.float)))

                    if not pd.isna(row['SugarContent']):
                        g.add((nutrition_info_uri, SDO.sugarContent, Literal(row['SugarContent'], datatype=XSD.float)))

                    if not pd.isna(row['ProteinContent']):
                        g.add((nutrition_info_uri, SDO.proteinContent,
                               Literal(row['ProteinContent'], datatype=XSD.float)))

                    # add description to dish
                    if not pd.isna(row['Description']):
                        g.add((dish_uri, SDO.description, Literal(row['Description'], datatype=XSD.string)))

            # create recipeId TYPE
            recipe_uri = ex['recipe_' + str(row["RecipeId"])]
            g.add((recipe_uri, RDF.type, SDO.Recipe))
            # Name
            if not pd.isna(row['Name']):
                g.add((recipe_uri, SDO.name, Literal(row['Name'], datatype=XSD.string)))
                # add recipe as one of the recipes for a dish
                g.add((ex[row['Name']], ex.hasRecipe, recipe_uri))
            # cookTime
            if not pd.isna(row['CookTime']):
                g.add((recipe_uri, SDO.cookTime, Literal(row['CookTime'], datatype=XSD.duration)))
            # prepTime
            if not pd.isna(row['PrepTime']):
                g.add((recipe_uri, SDO.prepTime, Literal(row['PrepTime'], datatype=XSD.duration)))
            # datePublished
            if not pd.isna(row['DatePublished']):
                g.add((recipe_uri, SDO.datePublished,
                       Literal(row['DatePublished'], datatype=XSD.dateTime)))
            # images
            if not pd.isna(row['Images']):
                g.add((recipe_uri, SDO.image, Literal(row['Images'], datatype=XSD.anyURI)))
            # recipecategory
            if not pd.isna(row['RecipeCategory']):
                g.add((recipe_uri, SDO.recipeCategory,
                       Literal(row['RecipeCategory'], datatype=XSD.string)))
                # Keywords
            if not pd.isna(row['Keywords']).any():
                for keyword in row['Keywords']:
                    g.add(
                        (recipe_uri, SDO.keywords, Literal(keyword, datatype=XSD.string)))
            # recipeYield
            if not pd.isna(row['RecipeYield']):
                g.add((recipe_uri, SDO.recipeYield, Literal(row['RecipeYield'], datatype=XSD.string)))
            # RecipeInstructions  
            if not pd.isna(row['RecipeInstructions']):
                g.add((recipe_uri, SDO.recipeInstructions,
                       Literal(row['RecipeInstructions'], datatype=XSD.string)))
            # Ingredients
            if not pd.isna(row['RecipeIngredientParts']).any():
                for ingredients in row['RecipeIngredientParts']:
                    ingredient = safe_uri_local(ingredients)
                    ingredient_uri = ex['Ingredient_' + ingredient]  # Use the prefixed string here
                    if ingredient not in ingredient_instances:
                        ingredient_instances[ingredient] = ingredient_uri  # Record it if necessary
                        g.add((ingredient_uri, RDF.type, ex['Ingredient']))
                        g.add((ingredient_uri, RDFS.label, Literal(ingredients, datatype=XSD.string)))
                    g.add((recipe_uri, ex['containsIngredient'], ingredient_uri))


    if reviews is not None:
        for index, row in reviews.iloc[:amount].iterrows():
            review_uri = ex["review_" + str(row['ReviewId'])]
            g.add((review_uri, RDF.type, SDO.Review))
            g.add((ex['recipe_' + str(row["RecipeId"])], SDO.review, review_uri))
            g.add(
                (review_uri, SDO.datePublished, Literal(row['DateSubmitted'], datatype=XSD.dateTime)))
            g.add((review_uri, SDO.dateModified, Literal(row['DateModified'], datatype=XSD.dateTime)))
            g.add((review_uri, SDO.reviewBody, Literal(row['Review'], datatype=XSD.string)))

            # add author and person
            # I use alternateName here since its an alias.
            author_uri = ex["author_" + str(row['AuthorId'])]
            g.add((author_uri, RDF.type, SDO.Person))
            g.add((author_uri, SDO.alternateName, Literal(row['AuthorName'], datatype=XSD.string)))

            # now we can add the author of review
            g.add((review_uri, SDO.author, author_uri))

    if restaurants is not None:
        for index, row in restaurants.iloc[:amount].iterrows():
            restaurant_uri = ex['restaurant_' + str(row['Restaurant ID'])]
            g.add((restaurant_uri, RDF.type, SDO.Restaurant))

            if not pd.isna(row['Restaurant Name']):
                g.add((restaurant_uri, SDO.legalName,
                       Literal(row['Restaurant Name'], datatype=XSD.string)))

            # Country
            if not pd.isna(row['Country']):
                g.add(
                    (restaurant_uri, ex.inCountry, Literal(row['Country'], datatype=XSD.string)))

            # City
            if not pd.isna(row['City']):
                g.add(
                    (restaurant_uri, SDO.areaServed, Literal(row['City'], datatype=XSD.string)))

            # address
            if not pd.isna(row['Address']):
                g.add(
                    (restaurant_uri, SDO.address, Literal(row['Address'], datatype=XSD.string)))

            # locality
            if not pd.isna(row['Locality Verbose']):
                g.add((restaurant_uri, SDO.location,
                       Literal(row['Locality Verbose'], datatype=XSD.string)))

            # longitude
            if not pd.isna(row['Longitude']):
                g.add((restaurant_uri, SDO.longitude,
                       Literal(row['Longitude'], datatype=XSD.float)))

            # latitude
            if not pd.isna(row['Latitude']):
                g.add(
                    (restaurant_uri, SDO.latitude, Literal(row['Latitude'], datatype=XSD.float)))

            # cuisines
            if not pd.isna(row['Cuisine_List']).any():
                for cuisine in row['Cuisine_List']:
                    cuisine = sanitize_name(cuisine)
                    cuisine_uri = ex['is_' + cuisine]
                    g.add((cuisine_uri, RDF.type, ex.Cuisine))
                    g.add((restaurant_uri, ex.hasCuisine, cuisine_uri))
                    g.add((cuisine_uri, RDFS.label, Literal(cuisine, datatype=XSD.string)))

            # currencies
            if not pd.isna(row['Currency']):
                g.add((restaurant_uri, SDO.currenciesAccepted,
                       Literal(row['Currency'], datatype=XSD.string)))

            # tableBooking
            if not pd.isna(row['Has Table booking']):
                g.add((restaurant_uri, ex.hasTableBooking,
                       Literal(bool(row['Has Table booking']), datatype=XSD.boolean)))

            # online delivery
            if not pd.isna(row['Has Online delivery']):
                g.add((restaurant_uri, ex.hasDelivery,
                       Literal(bool(row['Has Online delivery']), datatype=XSD.boolean)))

            # Is delivering now
            if not pd.isna(row['Is delivering now']):
                g.add((restaurant_uri, ex.isDeliveringNow,
                       Literal(bool(row['Is delivering now']), datatype=XSD.boolean)))

            # Switch to order menu
            if not pd.isna(row['Switch to order menu']):
                g.add((restaurant_uri, ex.switchToOrderMenu,
                       Literal(bool(row['Switch to order menu']), datatype=XSD.boolean)))

            # priceRange
            if not pd.isna(row['Average cost of two in USD']):
                g.add((restaurant_uri, SDO.priceRange,
                       Literal(str(row['Average cost of two in USD']), datatype=XSD.string)))

            # aggregate rating
            if not pd.isna(row['Aggregate rating']):
                g.add((restaurant_uri, ex.avgRating,
                       Literal(float(row['Aggregate rating']), datatype=XSD.double)))

            # votes
            if not pd.isna(row['Votes']):
                g.add(
                    (restaurant_uri, ex.votes, Literal(int(row['Votes']), datatype=XSD.integer)))

            # number of dishes
            if not pd.isna(row['Nummber of dishes in cuisines']):
                g.add((restaurant_uri, ex.numberOfDishes,
                       Literal(int(row['Nummber of dishes in cuisines']), datatype=XSD.integer)))


```
